```python
from pathlib import Path

import pandas as pd
import rich
```

```python
allseriespath = Path("allseries.tsv")
df = pd.read_csv(allseriespath, sep="\t")
```

```python
df
```

```python
# Find StudyInstanceUIDs that contain both 'PT' and 'CT'
valid_studies = df.groupby("StudyInstanceUID")["Modality"]\
	.transform(lambda x: {"PT", "CT"}.issubset(set(x)))

df_ptct = df[valid_studies]
df_ptct
```

```python
# use the valid studies to filter the original dataframe
# and count the number of CTs and PTs in each study
studyptct = df[df["StudyInstanceUID"].isin(df_ptct.StudyInstanceUID.unique())] \
    .groupby(["StudyInstanceUID", "Modality"]) \
    .size() \
    .unstack(fill_value=0) \
    [['CT', 'PT', 'Collection']]

# order by decreasing number of CTs and then PTs
studyptct = studyptct.sort_values(['CT', 'PT'], ascending=False)

studyptct.to_csv("Study-Pet-CT_modality-counts.tsv", sep="\t")

# count unique combos of CT and PT
studyptct['combo'] = studyptct.apply(lambda x: f"{x['CT']} CT, {x['PT']} PT", axis=1)

studyptct = studyptct.groupby('combo').size().reset_index(name='count')
studyptct = studyptct.sort_values(['combo'], key=lambda x: x.str.extract(r'(\d+) CT, (\d+) PT').astype(int).sum(axis=1), ascending=False)
# reverse order by combo name
# studyptct = studyptct.iloc[::-1]

studyptct.to_csv("Study-Pet-CT_modality-combos.tsv", sep="\t", index=False)

```
